```python
import pandas as pd

# Read the data
data = pd.read_csv('data/acde_person.csv')
top500 = pd.read_csv('data/ACDE_DAAO_Biography_Selected_v_1_1_202211101335.csv')
```

```python
import numpy as np

daao_persons = data[data['data_source'].str.contains('DAAO')].copy()
daao_persons['ori_url2'] = daao_persons['ori_url'].apply(lambda x: eval(x))

top500_df = pd.DataFrame()

for i in top500[top500['Expert_Checked']]['ori_url']: top500_df = top500_df.append(daao_persons[daao_persons['ori_url2'] == i])

# remove last column of the dataframe
top500_df = top500_df.iloc[:, :-1]
top500_df.head()
```

```python
# output as xlsx
top500_df.to_excel('data/daao500_may2023.xlsx', index=False)
```
